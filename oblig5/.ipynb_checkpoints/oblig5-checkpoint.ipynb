```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('agaricus-lepiota.csv')

#Exploring data
habitat = df.loc[:,["edibility", "habitat"]]
e_habitat = habitat[habitat['edibility'] == 'e'].groupby("habitat")
p_habitat = habitat[habitat['edibility'] == 'p'].groupby("habitat")

print(e_habitat.size())
print(p_habitat.size())

labels1 = ['d','g','l','m','p','u','w']
labels2 = ['d','g','l','m','p','u']

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Habitat distributed between edibility vs non-edibility')
ax1.set_ylim([0, 2300])
ax2.set_ylim([0, 2300])
ax1.bar(labels1, e_habitat.size().to_numpy())
ax2.bar(labels2, p_habitat.size().to_numpy())
```
